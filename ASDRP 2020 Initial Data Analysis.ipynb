```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('GendervIndustry.csv')
```

```python
men_column19 = df['Unemployment Rate of Men June 2019']
men_column20 = df['Unemployment Rate of Men June 2020']

change_men20 = []
change_men19 = []

for x in range(0, len(men_column19)):
    if men_column20[x] > men_column19[x]:
        change_men20.append(men_column20[x]-men_column19[x])
        change_men19.append(None)
    else:
        change_men19.append(men_column19[x]-men_column20[x])
        change_men20.append(None)

df['Increase in Men Unemployment Rate 2020'] = change_men20
df['Decrease in Men Unemployment Rate 2020'] = change_men19

print(df.head(25))
```

```python
y = np.arange(df['Industry and class of worker'].size)

fig, axes = plt.subplots(ncols=2, sharey=True)

axes[0].barh(y, df['Decrease in Men Unemployment Rate 2020'], align='center', color="blue", zorder=10)
axes[0].set(title='Decrease in Men Unemployment Rate 2020')
axes[1].barh(y, df['Increase in Men Unemployment Rate 2020'], align='center', color="blue", zorder=10)
axes[1].set(title='Increase in Men Unemployment Rate 2020')

axes[0].invert_xaxis()
axes[0].set(yticks=y, yticklabels=df['Industry and class of worker'])
axes[0].yaxis.tick_right()

for ax in axes.flat:
    ax.margins(0.03)
    ax.grid(True)
    
fig.subplots_adjust(wspace=0.73)
fig = plt.gcf()
fig.set_size_inches(24,14)
plt.show()
```

```python
women_column19 = df['Unemployment Rate of Women June 2019']
women_column20 = df['Unemployment Rate of Women of June 2020']

change_women20 = []
change_women19 = []

for x in range(0, len(women_column19)):
    if women_column20[x] > women_column19[x]:
        change_women20.append(women_column20[x]-women_column19[x])
        change_women19.append(None)
    else:
        change_women19.append(women_column19[x]-women_column20[x])
        change_women20.append(None)

df['Increase in Women Unemployment Rate 2020'] = change_women20
df['Decrease in Women Unemployment Rate 2020'] = change_women19

print(df.head(25))
```
