```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv("stateovertime.csv", index_col = 'State')
```

```python
state_dict = {}

for state in data.columns:
    state_dict[state] = data[state]
```

```python
fig, a = plt.subplots(5, 10)
row = 0
column = 0

for state in state_dict.keys():    
    
    
    x = range(len(state_dict[state]))
    temp = state_dict[state]
    
    #plt.scatter(x, temp)
    my_model = np.poly1d(np.polyfit(x, temp,10))
    my_line = np.linspace(1, x, 100)
    
    
    a[row][column].plot(my_line, my_model(my_line))
    a[row][column].scatter(x, temp)
    a[row][column].set_title(state)
    
    fig = plt.gcf()
    fig.set_size_inches(20,12)
    fig.subplots_adjust(hspace=0.50)
    
    column += 1    
    if (column == 10 and row < 4):
        row += 1
        column = 0
    elif (row == 4 and column == 10):
        break
    
    

plt.show() 
plt.savefig("state_linear_regression_graphs.png")

#print(state_dict['Alabama'])
```
