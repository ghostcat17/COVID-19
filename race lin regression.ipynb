```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv("raceovertime.csv", index_col = 'Month')

white = data['White']
black = data['Black or African American']
asian = data['Asian']
latino = data['Hispanic or Latino']

x = range(len(white))

plt.scatter(x, white)
plt.scatter(x, black)
plt.scatter(x, asian)
plt.scatter(x, latino)

my_model = np.poly1d(np.polyfit(x, white, 10))
my_line = np.linspace(1, x, 100)

my_model2 = np.poly1d(np.polyfit(x, black,10))
my_line2 = np.linspace(1, x, 100)

my_model3 = np.poly1d(np.polyfit(x, asian,10))
my_line3 = np.linspace(1, x, 100)

my_model4 = np.poly1d(np.polyfit(x, latino,10))
my_line4 = np.linspace(1, x, 100)

plt.plot(my_line, my_model(my_line))
plt.plot(my_line2, my_model2(my_line2))
plt.plot(my_line3, my_model3(my_line3))
plt.plot(my_line4, my_model4(my_line4))

plt.show()
```
