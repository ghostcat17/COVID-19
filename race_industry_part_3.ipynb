```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df=pd.read_csv("Rates for Industry (Percentages) - Sheet1.csv")

# set width of bar
barWidth = 0.15
plt.figure(figsize=(12,8))
 
# set height of bar
manage = [0.87,-1.56,19.48,-22.74]
service = [41.85,51.48,44.15,54.05]
sales = [30.43,35.08,19.90,28.02]
construct=[14.39,2.24,1.68,34.59]
transp=[12.45,12.76,14.22,6.08]

men2019_means = [3.6, 3.1, 6.6, 2.2, 3.7]
women2019_means = [3.5, 3.1, 5.6, 2.2, 4.3]
men2020_means=[11.9,10.9,16.2,13.0,15.2]
women2020_means=[14.0,13.3,16.1,15.8,18.7]
 
# Set position of bar on X axis
r1 = np.arange(len(manage))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
r4 = [x + barWidth for x in r3]
r5 = [x + barWidth for x in r4]
# Make the plot

plt.bar(r1, manage, color='blue', width=barWidth, edgecolor='white', label='Management, professional, and related occupations')
plt.bar(r2, service, color='red', width=barWidth, edgecolor='white', label='Service occupations')
plt.bar(r3, sales, color='orange', width=barWidth, edgecolor='white', label='Sales and office occupations')
plt.bar(r4, construct, color='green', width=barWidth, edgecolor='white', label='Natural resources, construction, and maintenance occupations')
plt.bar(r5, transp, color='green', width=barWidth, edgecolor='white', label='Production, transportation, and material moving occupations')
 
# Add xticks on the middle of the group bars
plt.xlabel('Race', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(men2019_means))],['White', 'Black', 'Asian', 'Hispanic'])
plt.yticks([-25,0,25,50,75])
plt.ylabel("Change in unemployed persons (thousands)")
plt.title("Unemployment Rate by Industry")


#[r + barWidth for r in range(len(bars1))], 
# Create legend & Show graphic
plt.legend()
plt.savefig("Unemployment_Industry_Numbers.png")
plt.show()
```
