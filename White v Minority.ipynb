```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns; sns.set()
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv("Second Data Analysis without population CA,AZ,NV - Sheet1 (1).csv")
df=df.drop(["Counties"],axis=1)
df=df.drop(["% of county Black"],axis=1)
df=df.drop(["% of county Asian"],axis=1)
df=df.drop(["% of county Hispanic"],axis=1)
X=df.values
print(X)
```

```python
print(__doc__)


# Code source: Gal Varoquaux
# Modified for documentation by Jaques Grobler
# License: BSD 3 clause

import numpy as np
import matplotlib.pyplot as plt
# Though the following import is not directly being used, it is required
# for 3D projection to work
from mpl_toolkits.mplot3d import Axes3D

from sklearn.cluster import KMeans
from sklearn import datasets

np.random.seed(5)

iris = datasets.load_iris()
y = iris.target

estimators = [
              ('k_means_iris_3', KMeans(n_clusters=3)),
              ]

fignum = 1
titles = ['3 clusters']
for name, est in estimators:
    fig = plt.figure(fignum, figsize=(12, 8))
    ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=45)
    est.fit(X)
    labels = est.labels_

    ax.scatter(X[:, 1], X[:, 2], X[:, 0],
               c=labels.astype(np.float), edgecolor='k')

    ax.w_xaxis.set_ticklabels([])
    ax.w_yaxis.set_ticklabels([])
    ax.w_zaxis.set_ticklabels([])
    ax.set_xlabel('White %')
    ax.set_zlabel('Increase in Unemployment')
    ax.set_ylabel('Minority %')
    ax.set_title(titles[fignum - 1])
    ax.dist = 12
    fignum = fignum + 1

    print(est.cluster_centers_)

fig.show()
```
